```python
import csv
import pandas as pd

df = pd.read_csv ('Traffic_camera_offences_and_fines.csv')
# df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
```

```python
CameraDF = df.groupby(['Camera_Type'])['Sum_Inf_Count'].sum()
CameraDF.columns=['Camera_Type','CountofOffences']
CameraDF.plot(kind = 'bar', x = 'Camera_Type', color = ['Blue'])
plt.title('Count of traffic offences by detected Camera type')
plt.xlabel('Camera Type')
plt.ylabel('Count of offences')
```

```python
CategoryDF = df.groupby(['Clt_Catg'])['Sum_Inf_Count'].sum()
CategoryDF.columns=['Clt_Catg','CountofOffences']
CategoryDF.plot(kind = 'bar', x = 'Clt_Catg', color = ['Green'])
plt.title('Count of traffic offences conducted by Client type')
plt.xlabel('Client Category')
plt.ylabel('Count of offences')
```
